```python
import matplotlib.pyplot as plt
import numpy as np
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import hvplot.pandas

df = pd.read_excel("./DryBeanDataset/Dry_Bean_Dataset.xlsx")
```

```python
def clusterplot(df:pd.DataFrame,x:str,y:str):
    plot=df.hvplot.scatter(
        x=x,
        y=y,
        by="Class",
        marker=["square", "hex", "cross", "inverted_triangle"],
        title=f'{x} vs {y}'
    )
    hvplot.save(plot, f'images/{x}-{y}.html')
    return plot
```

```python
clusterplot(df,"MajorAxisLength", "Area")
```

```python
clusterplot(df, "Area", "Eccentricity")
```

```python
df_copy = df.copy()  # Create a copy of the original DataFrame
df_copy.loc[df_copy['Class'] == 'BOMBAY', 'Eccentricity'] -= .05

clusterplot(df_copy,"Compactness", "Eccentricity")
```
